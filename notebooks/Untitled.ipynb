```python
import matplotlib.pyplot as plt
%matplotlib inline
from IPython import display
```

```python
def show_state(env, step=0, info=""):
    plt.figure(3,figsize=[4,3])
    plt.clf()
    plt.imshow(env.render(mode='rgb_array'))
    #plt.title("%s | Step: %d %s" % (env._spec.id,step, info))
    plt.axis('off')

    display.clear_output(wait=True)
    display.display(plt.gcf())
```

```python
env=gym.make('AntBulletEnv-v0')
```

```python
%%time
#env = gym.make('CartPole-v0')
#env = gym.make('Pong-v0')
env.reset()
for _ in range(1000):
    #env.render()
    env.step(env.action_space.sample()) # take a random action
    show_state(env)
```

```python
env = gym.make('Breakout-ramDeterministic-v4')
```

```python
#env = gym.wrappers.Monitor(env, 'drive/My Drive/RL_movies/breakout',video_callable=lambda episode_id: True,force=True)
env = gym.wrappers.Monitor(env, 'mp4/breakout_test',video_callable=lambda episode_id: episode_id%100==0,force=True)
```

```python
%%time
env.reset()
print(env.action_space)
for episode in range(1000):
  state = env.reset()
  while True:
    random_action=env.action_space.sample()
    observation, reward, done, info = env.step(random_action) # take a random action
    if done:
        print('Finished episode: {} '.format(episode))
        break
env.close()
```
